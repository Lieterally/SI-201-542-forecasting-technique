```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.rcParams.update({
  'font.size': 12,
  'grid.alpha': 0.25})
```

```python
def get_exp_declining_weights(alpha, period):
  n = len(period)
  return alpha * (1 - alpha)**np.arange(n)
```

```python
fig, ax = plt.subplots()

n = 10

alpha = 0.1
period = -np.arange(n)[::-1]
print(period)
weight_arr = get_exp_declining_weights(alpha, period)[::-1]

ax.plot(period, weight_arr, marker='o', linestyle="None", 
  markerfacecolor="None", markeredgewidth=2, label=r"$\alpha$" + f" = {alpha}")

alpha = 0.6
weight_arr = get_exp_declining_weights(alpha, period)[::-1]

ax.plot(period, weight_arr, marker='x', linestyle="None",  
  markerfacecolor="None", markeredgewidth=2, label=r"$\alpha$" + f" = {alpha}")

ax.set_xlabel(r"period, $k$")
ax.set_ylabel(r"$\alpha(1-\alpha)^{|k|}$")
ax.set_title("exponential declining weights")
ax.legend(loc="upper left")
ax.grid("on")
plt.show(fig)
```
